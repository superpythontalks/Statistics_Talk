You have two random variables A and B. A is a beta distribution with a=2, b=2, loc=0, scale=1. B is a Gaussian distribution with mean=0.2 and std=0.2. You can create a new random variable C = A - B. Take two samples of A and B of size 1000 each and subtract them to get a sample for C. Plot it's histogram. Does is look symmetric? Plot the PDFs of A and B on the same plot over the outcomes in [0, 2]. Can you explain the lack of symmetry of the histogram of C from this plot?

```python
% matplotlib inline
from scipy import stats
import numpy as np
from matplotlib import pyplot as plt

rvA = stats.beta(2, 2)
rvB = stats.norm(loc=0.2, scale=0.2)

sampleA = rvA.rvs(size=1000)
sampleB = rvB.rvs(size=1000)
sampleC = sampleA - sampleB

fig, axs = plt.subplots(nrows=2, figsize=(10, 8))

x = np.linspace(0, 2, 500)
axs[0].hist(sampleC, normed=True)
axs[0].set_ylabel('Relative Frequency')
axs[0].set_xlabel('outcome of C')
axs[1].plot(x, rvA.pdf(x), 'b-', label='rvA')
axs[1].plot(x, rvB.pdf(x), 'r-', label='rvB')
axs[1].set_ylabel('PDF')
axs[1].set_xlabel('outcome')
axs[1].legend()
```
